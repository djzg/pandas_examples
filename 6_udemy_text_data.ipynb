```python
import pandas as pd
```

```python
chicago = pd.read_csv('chicago.csv').dropna(how='all')
chicago.head(3)
```

```python
chicago.info()
```

```python
chicago.nunique()
```

```python
chicago['Department'] = chicago['Department'].astype('category')
```

```python
chicago.info()
```

```python
## .lower(), .upper(), .title(), .len() methods

'HELLO WORLD'.lower()
'hello world'.upper()
'hello world'.title()
```

```python
chicago['Name'] = chicago['Name'].str.title() # .str is necessary to avoid error
chicago['Position Title'] = chicago['Position Title'].str.title()
```

```python
chicago.head(3)
```

```python
chicago['Department'].str.len()
```

```python
# .replace() method
chicago['Department'] = chicago['Department'].str.replace('MGMNT', 'MANAGEMENT')
```

```python
chicago.head(3)
```

```python
chicago['Employee Annual Salary'] = chicago['Employee Annual Salary'].str.replace('$', '')
```

```python
chicago['Employee Annual Salary'] = chicago['Employee Annual Salary'].astype('float')
```

```python
chicago['Employee Annual Salary'].describe()
```

```python
## Filtering with string methods

mask = chicago['Position Title'].str.lower().str.contains('water') # normalizing values to lowercase for easier searching 
```

```python
chicago[mask]
```

```python
mask2 = chicago['Position Title'].str.lower().str.startswith('water')
```

```python
chicago[mask2]
```

```python
mask3 = chicago['Position Title'].str.lower().str.endswith('ist')
```

```python
chicago[mask3]
```

```python
## .strip(), .lstrip() and .rstrip() methods

chicago['Name'].str.lstrip()
chicago['Name'].str.rstrip()
chicago['Name'] = chicago['Name'].str.strip()
```

```python
chicago['Position Title'] = chicago['Position Title'].str.lstrip().str.rstrip()
```

```python
## String methods on index and columns
chicago = pd.read_csv('chicago.csv', index_col='Name').dropna(how='all')
chicago['Department'] = chicago['Department'].astype('category')
chicago.tail(3)
```

```python
chicago.index = chicago.index.str.strip().str.title()
```

```python
chicago.head(3)
```

```python
chicago.columns = chicago.columns.str.upper()
```

```python
chicago.head(3)
```

```python
chicago = pd.read_csv('chicago.csv').dropna(how='all')
chicago['Department'] = chicago['Department'].astype('category')
chicago.tail(3)
```

```python
## .str.split() method

chicago['First Name'] = chicago['Name'].str.split(",").str.get(0).str.title()
chicago.head(4)
```

```python
chicago['Position Title'].str.split(' ').str.get(0).value_counts()
```

```python
chicago['Name'].str.split(',').str.get(1).str.strip().str.split(' ').str.get(0).value_counts()
```

```python
## expand and n parameters of the .str.split() method

chicago[['First Name', 'Last Name']] = chicago['Name'].str.split(',', expand=True) # returns dataframe instead of a list
```

```python
chicago
```

```python
chicago['Position Title'].str.split(' ', expand=True)
```

```python
chicago[['First Title Word', 'Remaining Word']] = chicago['Position Title'].str.split(' ', expand=True, n=1) # n - number of splits
```
