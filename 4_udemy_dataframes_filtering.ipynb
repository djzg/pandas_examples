```python
# using parse_date
df = pd.read_csv('employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df.head(3)
```

## Filtering a dataframe based on condition

```python
df[df['Gender'] == 'Male']
```

```python
mask = df['Team'] == 'Finance'
df[mask]
```

```python
df[df['Senior Management']]
```

```python
df[df['Team'] != 'Marketing']
```

```python
df[df['Salary'] > 110000]
```

```python
df[df['Start Date'] <= '1985-01-01']
```

```python
mask1 = df['Gender'] == 'Male'
mask2 = df['Team'] == 'Marketing'
df[mask1 & mask2].head()
```

```python
df[mask1 | mask2]
```

```python
mask3 = df['First Name'] == 'Robert'
mask4 = df['Team'] == 'Client Services'
mask5 = df['Start Date'] > '2016-06-01'

df[(mask3 & mask4) | mask5]
```

```python
## .isin() method

mask = df['Team'].isin(['Legal', 'Sales', 'Product'])
df[mask]
```

```python
## .isnull() and .notnull() methods

mask = df['Team'].isnull()
df[mask].head()
```

```python
not_null_mask = df['Team'].notnull()
df[not_null_mask].head()
```

```python
## .between() method

between_mask = df['Salary'].between(60000, 70000)
df[between_mask]
```

```python
df[df['Bonus %'].between(2.0, 5.0)]
```

```python
df[df['Start Date'].between('1991-01-01', '1992-01-01')].head()
```

```python
df[df['Last Login Time'].between('11:50AM', '12:30PM')]
```

```python
## .duplicated() method
df.sort_values('First Name', inplace=True)
df.head(3)
```

```python
df[df['First Name'].duplicated()]
```

```python
df[df['First Name'].duplicated(keep='last')]
```

```python
df[df['First Name'].duplicated(keep=False)]
```

```python
df[~df['First Name'].duplicated(keep=False)] # keeping only unique records using tilde
```

```python
## .drop_duplicated() method

print(len(df))
print(len(df.drop_duplicates()))
print(len(df.drop_duplicates(subset=['First Name'], keep='first')))
```

```python
df.drop_duplicates(subset=['First Name'], keep='first')
```

```python
df.drop_duplicates(subset=['First Name'], keep=False)
```
