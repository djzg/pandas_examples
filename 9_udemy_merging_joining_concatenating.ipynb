```python
import pandas as pd
```

```python
week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customer = pd.read_csv('Restaurant - Customers.csv')
food = pd.read_csv('Restaurant - Foods.csv')
```

```python
food
```

```python
## .concat() method

week1.head(3)
```

```python
week2.head(3)
```

```python
pd.concat(objs=[week1, week2])
pd.concat(objs=[week1, week2], ignore_index=True)
```

```python
week1.append(week2, ignore_index=True)
```

```python
pd.concat(objs=[week1, week2], keys=['Week 1', 'Week2'])
```

```python
sales = pd.concat(objs=[week1, week2], keys=['Week 1', 'Week2'])
```

```python
## Inner Joins
week1.head(3)
```

```python
week2.head(3)
```

```python
week1.merge(week2, how='inner', on='Customer ID', suffixes=[' - week 1', ' - week 2'])
```

```python
week1.merge(week2, how='inner', on=['Customer ID', 'Food ID'])
```

```python
## Outer joins

print(week1.head(3))
print(week2.head(3))
```

```python
week1.merge(week2, how='outer', on='Customer ID', suffixes=[' - Week 1', ' - Week 2'])
```

```python
merged = week1.merge(week2, how='outer', on='Customer ID', suffixes=[' - Week 1', ' - Week 2'], indicator=True)
merged
```

```python
merged['_merge'].value_counts()
```

```python
mask = merged['_merge'].isin(['left_only', 'right_only'])
```

```python
merged[mask]
```

```python
## Left joins

print(week1.head(3))
print(food.head(3))
```

```python
week1.merge(food, how='left', on='Food ID', sort=True)
```

```python
## left_on and right_on parameters

print(week2.head(3))
print(customer.head(3))
```

```python
week2.merge(customer, how='left', left_on='Customer ID', right_on='ID', indicator=True)
```

```python
week2.merge(customer, how='left', left_on='Customer ID', right_on='ID', indicator=True, sort=True).drop('ID', axis='columns')
```

```python
## Merging by indexes with the left_index and right_index parameters

customer.set_index('ID', inplace=True)
```

```python
food.set_index('Food ID', inplace=True)
```

```python
print(customer.head(3))
print(food.head(3))
```

```python
sales = week1.merge(customer, how='left', left_on='Customer ID', right_index=True)
sales
```

```python
final_sales = sales.merge(food, how='left', left_on='Food ID', right_index=True)
final_sales
```

```python
week1.merge(week2, how='left', left_index=True, right_index=True, indicator=True, suffixes=[' - Week 1', ' - Week 2'])
```
