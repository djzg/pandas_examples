```python
## .stack() method

world = pd.read_csv('worldstats.csv', index_col=['country', 'year'])
world
```

```python
stacked = world.stack()
stacked
```

```python
## .unstack() method

stacked.unstack()
```

```python
stacked.unstack().unstack()
```

```python
stacked.unstack().unstack().unstack()
```

```python
stacked.unstack(0)
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(-1) # same as .unstack(2)
```

```python
stacked.unstack('country')
```

```python
stacked.unstack(level=[1,2])
```

```python
stacked.unstack(level=[1, 0])
```

```python
stacked.unstack(level=['country',2])
```

```python
stacked.unstack('year', fill_value=0)
```

```python
## .pivot() method

salesmen = pd.read_csv('salesmen.csv', parse_dates=['Date'])
salesmen['Salesman'] = salesmen['Salesman'].astype('category')
salesmen
```

```python
salesmen.pivot(index='Date', columns='Salesman', values='Revenue')
```

```python
## .pivot_table() method

foods = pd.read_csv('foods.csv')
foods
```

```python
foods.pivot_table(values='Spend', index='Gender', aggfunc='mean')
```

```python
foods.pivot_table(values='Spend', index='Gender', aggfunc='sum')
```

```python
foods.pivot_table(values='Spend', index='First Name', aggfunc='sum')
```

```python
foods.pivot_table(values='Spend', index='Item', aggfunc='sum')
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], aggfunc='mean')
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], columns='City', aggfunc='sum')
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], columns=['Frequency', 'City'], aggfunc='sum')
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], columns='City', aggfunc='count')
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], columns='City', aggfunc='max')
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], columns='City', aggfunc='min')
```

```python
pd.pivot_table(data=foods, values='Spend', index=['Gender', 'Item'], columns='City', aggfunc='max')
```
